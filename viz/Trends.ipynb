```python
def infer_dtype(val):
    """
    Using regex, infer a limited number of dtypes for string `val`
    (only dtypes expected to be found in a Google Trends CSV export).
    """
    if re.match(r'\d{4}-\d{2}(?:-\d{2})?', val):
        return 'date'
    elif re.match(r'[+-]?\d+$', val):
        return 'int'
    elif re.match("[+-]\d+\s*%?", val):
        return 'text'
    elif re.match(r'[+-]?\d+%$', val):
        return 'pct'
    elif re.match(r'[a-zA-Z ]+', val):
        return 'text'
    else:
        msg = "val={0} dtype not recognized".format(val)
        return "text"
        #raise ValueError(msg)
        
pytrends._infer_dtype = infer_dtype
```

```python
translations = pd.read_csv("/home/rmyeid/notebooks/Translation Calculator - Human Trafficking.tsv", delimiter="\t")
translations = translations[["Language Code", "Language Name", "Unnamed: 4"]]
translations.rename(columns={"Unnamed: 4": "Translations"}, inplace=True)
```

```python
dfs = {}

for code, lang, term in translations.values:
    code = code.strip()
    lang = lang.strip()
    if len(code.strip()) > 2: continue
    try:
      connector.request_report(term)
    except UnicodeEncodeError:
      print("We could not process language {}".format(lang))
      continue
    result = pytrends.parse_data(connector.get_data())
    if not result:
        print("Language {} is not found".format(lang))
        continue
    report = result
    df = pd.DataFrame.from_records([x for x in report.values() if len(x) > 300][0])
    week_column_name = df.dtypes.index[[k for k in df.dtypes.values  == "datetime64[ns]"]].values[0]
    df.rename(columns={week_column_name: "Week"}, inplace=True)
    df.set_index("Week", inplace=True)
    dfs[(code, lang)] = df
```

```python
def MA(x, N):
  return np.convolve(x, np.ones((N,))/N, mode='v')
```
